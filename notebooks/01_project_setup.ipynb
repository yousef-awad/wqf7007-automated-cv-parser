## Project Overview

This project aims to convert unstructured CV text into structured candidate records that are easier to search, compare, and analyze.

**Planned extraction targets**
- Job titles
- Technical skills
- Educational background


## Problem Statement

Resumes are unstructured and vary widely in format, making manual extraction of important candidate information slow, inconsistent, and labor-intensive for recruiters.


## Objectives

- Build an NLP pipeline to process raw CV text
- Automatically extract job titles, technical skills, and educational background
- Convert resume text into structured candidate records

## Planned NLP Pipeline

1. Ingest raw CV files or resume text
2. Extract machine-readable text from PDF or text sources
3. Clean and normalize the text
4. Tokenize the content for sequence labeling
5. Extract target entities such as job titles, skills, and education
6. Convert extracted outputs into structured candidate records


## Baseline Model Idea

**Task framing:** token classification / named entity recognition (NER)

**Baseline direction:** BERT-based sequence labeling

**Initial tag idea**
- `O`
- `B-JOB_TITLE`
- `I-JOB_TITLE`
- `B-SKILL`
- `I-SKILL`
- `B-EDUCATION`
- `I-EDUCATION`

This is a starting point only. Final labels depend on the usable structure of the chosen dataset.
